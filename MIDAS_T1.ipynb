```python
def get_all_tweets(name, api):
    tweets = []
    new_tweets = api.user_timeline(screen_name = name, count= 200)
    tweets.extend(new_tweets)
    old_tweet = tweets[-1].id - 1
    count = 0
    while len(new_tweets) > 0:
        new_tweets = api.user_timeline(screen_name=name,count=200, max_id = old_tweet)
        tweets.extend(new_tweets)
        count+=len(tweets)
        old_tweet = tweets[-1].id - 1
        for tweet in tweets:
            keys[tweet.id_str] = {}
            temp = {}
            temp["CREATED_AT"] = str(tweet.created_at)
            temp["TEXT"] = tweet.text.encode('utf-8')
            temp["FAVORITE_COUNT"] = tweet.favorite_count
            temp["RETWEET_COUNT"] = tweet.retweet_count
#             if 'media' in tweet.entities:
#                 if([medium['type'] == 'photo' for medium in tweet.entities['media']]):
#                 ct = len([medium['type'] == 'photo' for medium in tweet.entities['media']])
#             else:
#                 ct = 0
#             temp["COUNT_IMAGES"] = ct
            keys[tweet.id_str] = temp
    print('Number of tweets downloaded = %s' % len(tweets))
    with open('tweets_info.json', 'w') as data:
        json_string = json.dump(keys, data, indent=4, sort_keys=True)
    return json_string
```

```python
auth = tweepy.OAuthHandler(API_key, Secret_key)
auth.set_access_token(Access_token, Access_token_secret)
api = tweepy.API(auth)
handle = '@midasIIITD'
json_string = get_all_tweets(handle, api)
```
